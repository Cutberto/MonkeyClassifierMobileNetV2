Cargamos cualquier imagen que queramos analizar en el modelo y recibiremos la predicci贸n.

```python
#Seleccionamos una imagen del dataset de test para probar =) Aqu铆 se puede insertar cualquier imagen 
image = load_img('RUTA_DE_IMAGEN', target_size=(224, 224))

#EJEMPLO:
#image = load_img('C:/Users/cutic/OneDrive/Escritorio/Tecnol贸gico de Monterrey/Septimo semestre/DeepLearning/test/cabeza.jpeg', target_size=(224, 224))

# Pre-procesamos la imagen para que tenga las caracter铆sticas que el modelo requiere recibir
img = np.array(image)
img = img / 255.0
img = img.reshape(1,224,224,3)

# Realizamos la predicci贸n con el modelo
label = model.predict(img)

# Obtenemos el 铆ndice del valor m谩s grande del array de predicciones.
index = np.argmax(label)

# Usamos un diccionario para mapear el arreglo de predicciones con sus respectivas etiquetas
macacos = {0:"Mantled Howler", 1: "Patas Monkey", 2: "Bald Ukari", 3:"Japanese monjey", 4:"Pygmy marmoset", 5:"White Capucchino", 6: "Silverly marmoset", 7: "Spider Monkey", 8: "Black Headed Night Monkey", 9: "Nilgiri Langur"}

# Usamos una gr谩fica para mostrar la imagen analizada y el resultado del modelo
plt.imshow(image)
plt.axis("off")
plt.title("This monkey is a %s" %(macacos[index]))
```
